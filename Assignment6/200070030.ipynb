```python
import pandas as pd
import numpy as np
import sklearn.metrics as sk
import sklearn.linear_model as sm
from sklearn.model_selection import train_test_split
```

# **Q1: LINEAR REGRESSION**

```python
df = pd.read_excel("Real estate valuation data set.xlsx")
df.drop('No',axis=1,inplace=True)
df
```

Data after pre processing

```python
def regression(df_og,split,lambda_r=0,show_coeff=True,type='Ridge'): #Function to compute the regression coeffients
  df=df_og.copy()

  train, test = train_test_split(df, test_size=split, random_state=2)

  df=train.copy()
  df.drop("Y house price of unit area",axis=1,inplace=True)

  if lambda_r==0:
    model=sm.LinearRegression().fit(df,train["Y house price of unit area"])
  else:
    if type=='Ridge':
      model=sm.Ridge(alpha=lambda_r).fit(df,train["Y house price of unit area"])
    else:
      model=sm.Lasso(alpha=lambda_r).fit(df,train["Y house price of unit area"])

  if show_coeff:
    print("Optimal coefficients for split of",split*100,":",100-100*split)
    print("Intercept:",model.intercept_)
    j=0
    for i in model.coef_:
        print(df_og.columns[j],": ",i)
        j+=1
    print()
  else:
    if lambda_r>0:
      print(type)
    print("Observations with split of",split*100,":",100-100*split,",for ƛ=",lambda_r)

  df=df_og.copy()
  df.drop("Y house price of unit area",axis=1,inplace=True)
  predictions=model.predict(df)

  mse=sk.mean_squared_error(predictions,df_og["Y house price of unit area"])
  r_squared=model.score(df,df_og["Y house price of unit area"])

  print("Mean squared error for this solution:",mse)
  print("R² value for this solution:",r_squared)

  print()

  return predictions,mse,r_squared
```

The above function calculates the coefficients of a linear solution for the problem as well as the MSE(Mean Squared Error) and the Coefficient of Determination(R²)

```python
predictions,mse,r=regression(df,0.8)

splits=[0.6,0.7,0.9]
```

The coefficients and intercept for the split 80:20 for train and test data

```python
pd.set_option('display.max_rows', 500)
pd.DataFrame({'Predicted values':predictions,'Actual values':df['Y house price of unit area']})
```

Comparision of predicted values and actual values for the full dataset

```python
pd.reset_option('display.max_rows')
for split in splits:
  predictions,mse,r=regression(df,split)
  display(pd.DataFrame({'Predicted values':predictions,'Actual values':df['Y house price of unit area']}))
```

The required observations for splits of 60:40, 70:30 and 90:10

```python
splits=[0.6,0.7,0.8,0.9]
alphas=[0.001,0.005,0.01,0.05,0.1,0.5]
alpha_s=[0,0.001,0.001,0.005,0.005,0.01,0.01,0.05,0.05,0.1,0.1,0.5,0.5]
type_s=['Linear','Ridge','Lasso','Ridge','Lasso','Ridge','Lasso','Ridge','Lasso','Ridge','Lasso','Ridge','Lasso']
mse_s=[]
r_2_s=[]
for split in splits:
  print("SPLIT:",100*split,":",100-100*split)
  prediction,mse,r_2=regression(df,split,show_coeff=False)
  mses=[mse]
  r_2s=[r_2]
  for j in alphas:
    prediction,mse,r_2=regression(df,split,j,False)
    mses.append(mse)
    r_2s.append(r_2)
    prediction,mse,r_2=regression(df,split,j,False,'Lasso')
    mses.append(mse)
    r_2s.append(r_2)
  mse_s.append(mses)
  r_2_s.append(r_2s)
```

Observations for ridge regression with regularization parameter ƛ = 0.001,0.005,0.01,0.05,0.1,0.5

```python
for i in range(len(splits)):
  print("SPLIT:",100*splits[i],":",100-100*splits[i])
  display(pd.DataFrame({'Type':type_s,'Alpha':alpha_s,'MSE':mse_s[i],'R²':r_2_s[i]}))
  print()
```

# **Q1: LOGISTIC REGRESSION**

```python
df=pd.read_csv("haberman.data")
df=df.rename(columns={'30':'X1:age','64':'X2:year','1':'X3:no of nodes','1.1':'Y:survival'})
df['Y:survival']=df['Y:survival'].map({1:1,2:0})
df
```

Displaying the data after pre processing 


```python
def log_regression(df_og,split,show_table=True): #Function to compute the regression coeffients
  df=df_og.copy()

  train, test = train_test_split(df, test_size=split, random_state=2)

  df=train.copy()
  df.drop('Y:survival',axis=1,inplace=True)

  model=sm.LogisticRegression(random_state=0).fit(df,train['Y:survival'])

  
  print("Optimal coefficients for split of",split*100,":",100-100*split)
  print("Intercept:",model.intercept_[0])
  j=0
  for i in model.coef_[0]:
    print(df_og.columns[j],": ",i)
    j+=1

  df=test.copy()
  df.drop('Y:survival',axis=1,inplace=True)
  predictions=model.predict(df)

  if show_table:
    pd.set_option('display.max_rows', 100)
  else:
    pd.set_option('display.max_rows',10)

  display(pd.DataFrame({'Predicted values':predictions,'Actual values':test['Y:survival']}).reset_index(drop=True))
  
  accuracy = np.sum(test['Y:survival']==predictions)/len(test)
  classification_error=1-accuracy

  return accuracy,classification_error
```

Function to implement logistic regression and get the predictions for the test dataset and the accuracy and classification errors

```python
accuracy,classification_error=log_regression(df,0.8)
```

Optimal coefficients and predicted values vs actual values for a 80:20 split

```python
print("Accuracy:",accuracy)
print("Classification error:",classification_error)
```

Accuracy and classification error for 80:20 split

```python
pd.reset_option('display.max_rows')
accuracies=[accuracy]
class_errors=[classification_error]
split_s=["80:20"]
splits.remove(0.8)
for split in splits:
  split_str=str(int(100*split))+":"+str(int(100-int(100*split)))
  split_s.append(split_str)
  print("SPLIT:",split_str)
  accuracy,classification_error=log_regression(df,split,False)
  print("Accuracy:",accuracy)
  print("Classification error:",classification_error)
  print()
  accuracies.append(accuracy)
  class_errors.append(classification_error)
  
```

Observed accuracy, classification errors and the predictions for different possible splits of the data

```python
pd.DataFrame({'train:test':split_s,'Accuracy':accuracies,'Class errors':class_errors})
```
